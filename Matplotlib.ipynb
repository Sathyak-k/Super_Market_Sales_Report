```python
import matplotlib.pyplot as plt
x = ["Rice","Juice","Chololates","Snacks","Dall"]
y = [200,150,300,250,150]
plt.plot(x,y,color = "saddlebrown",marker = "D",linestyle = "-.")
plt.title("Sales")
plt.xlabel("Categories")
plt.ylabel("Rate")
plt.grid()
plt.show()
```

```python

plt.bar(x,y,color= "pink")
plt.title("Stock")
plt.xlabel("Categories")
plt.ylabel("Quantity")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
Juice = ["BeetJuice","BlueberryJuice","CranberryJuice","CarrotJuice","GrapeJuice"]
quantity = [50,60,25,35,45]
colors = ["deeppink","darkblue","crimson","darkorange","blueviolet"]
plt.pie(quantity,colors=colors,labels=Juice,autopct="%1.1f%%",startangle=140)
plt.axis("equal")
plt.show()
```
